## A Neural Algorithm of Artistic Style (翻译)

### 摘要

对于纯艺术, 尤其是绘画, 人们已经精通通过组合内容和图像风格之间的复杂作用来创造独特的视觉体验. 但是这部分的算法基础依然是未知的, 而且现有的人工系统中没有一个能够有这样类似的能力. 但是在视觉感知的一些其他主要领域中, 比如物体和面部识别, 一类叫做深度神经网络的受生物学启发的视觉模型最近展示出接近人类的性能. 在此我们提出一种基于深度神经网络的人工系统, 它能够创造出高感知质量的艺术风格图片. 该系统用神经表示来分离和重组任意图像的内容和风格, 提供一个神经算法来创造出艺术风格照片. 更进一步地, 鉴于生物视觉和我们这个性能最优的人工神经网络之间显著的相似性, 我们还提供了人类怎么去创造和感知艺术形象的一种思路.

### 正文

在图像处理任务中的一种强有力的深度神经网络叫做卷积神经网络. 卷积神经网络由很多层构成, 这些层是一些用前馈方式层次化地处理可视信息的微小计算单元. 每一层的这些单元可以看做是图片过滤器的集合, 每一个单元从输入图像中提取一个确定的特征. 因此, 由一些称之为特征映射的给定层的输出就是一些输入图像的差异过滤后的版本.

当CNN被训练作物体识别, 他们开发出一种图像的表示, 可以使得物体信息在层次化处理过程中显式的增长. 因此, 随着层次网络的处理过程, 输入图像转变成了逐渐越老越关注与图像的内容而不是具体的像素值的表示. 我们能通过只是用这一层的特征映射来重构图像进而直接可视化每一包含输入图像的层的信息(Fig 1, 内容重构, 阅读 Methods 部分进一步了解如何重构图像). 更高层次的神经网络层捕获的是依据输入图像中的物体和排列确定的高层次内容, 而不是限定于对每个确定的像素值得重构(Fig 1, 内容重构 d,e) 与之相反的, 低层次的层的重构就是简单的修改原始图像中的某一个确定的像素(Fig 1, 内容重构, a,b,c). 我们把高层次的网络层的特征称之为内容表示.

为了获得输入图像的风格表示, 我们使用一种最开始用于捕获纹理信息的一种特征空间. 这个特征空间在每一个网络层的过滤器的最顶上被构建出来. 通过加入很多层的特征相关性, 我们获得输入图像的静止的, 多尺度的表示, 这种表示捕获的是其纹理信息而不是全局排列方位信息.

Figure 1.

**CNN.** 一个给定的输入图像在 CNN 的每一个处理层次上表示为一组过滤后的图像. 随着层次处理过程中不同的过滤器的数量的逐渐增加, 通过一些降低取样的机制(e.g., 最大池化, 这种机制能够减少每一个神经层的单元数量), 过滤后的图像的大小会减少. 

**内容重构.** 我们能够通过只是用特定的层能够获取的网络层响应来重构图像, 这样就可以可视化 CNN 的每一个不同处理层次的信息. 我们从原始 VGG 网络的 a ~ e 层进行重构图像. 我们发现低层次的网络层重构出来的图像基本上比较完整(跟原始图像相比), 而在高层次的网络层, 具体的像素点有些就丢失了, 不过图像的高层次内容被保留了下来.

**风格重构.** 在原始 CNN 表示之上, 我们还构建了一种能够捕获输入图像的风格的新的特征空间. 风格表示通过计算 CNN 不同网络层中的不用特征之间的相关性. 我们从不同的 CNN 层子集的风格表示来重构输入图像的风格. 这可以在抛弃一些场景的全局排列的信息时以递增规模创建出匹配给定风格图像的风格的图像.

再次说明, 我们可以通过构建一张匹配给定风格图像的风格表示的图像来可视化的展现出这些构建于不同网络层的风格特征空间捕获的信息.
