## A Neural Algorithm of Artistic Style (翻译)

### 摘要

对于纯艺术, 尤其是绘画, 人们已经精通通过组合内容和图像风格之间的复杂作用来创造独特的视觉体验. 但是这部分的算法基础依然是未知的, 而且现有的人工系统中没有一个能够有这样类似的能力. 但是在视觉感知的一些其他主要领域中, 比如物体和面部识别, 一类叫做深度神经网络的受生物学启发的视觉模型最近展示出接近人类的性能. 在此我们提出一种基于深度神经网络的人工系统, 它能够创造出高感知质量的艺术风格图片. 该系统用神经表示来分离和重组任意图像的内容和风格, 提供一个神经算法来创造出艺术风格照片. 更进一步地, 鉴于生物视觉和我们这个性能最优的人工神经网络之间显著的相似性, 我们还提供了人类怎么去创造和感知艺术形象的一种思路.

### 正文

在图像处理任务中的一种强有力的深度神经网络叫做卷积神经网络. 卷积神经网络由很多层构成, 这些层是一些用前馈方式层次化地处理可视信息的微小计算单元. 每一层的这些单元可以看做是图片过滤器的集合, 每一个单元从输入图像中提取一个确定的特征. 因此, 由一些称之为特征映射的给定层的输出就是一些输入图像的差异过滤后的版本.

当CNN被训练作物体识别, 人们ß开发出一种图像的表示, 可以使得物体信息在层次化处理过程中显式的增长. 因此, 随着层次网络的处理过程, 输入图像转变成了逐渐越老越关注与图像的内容而不是具体的像素值的表示. 我们能通过只是用这一层的特征映射来重构图像进而直接可视化每一包含输入图像的层的信息(Fig 1, 内容重构, 阅读 Methods 部分进一步了解如何重构图像). 更高层次的神经网络层捕获的是依据输入图像中的物体和排列确定的高层次内容, 而不是限定于对每个确定的像素值得重构(Fig 1, 内容重构 d,e) 与之相反的, 低层次的层的重构就是简单的修改原始图像中的某一个确定的像素(Fig 1, 内容重构, a,b,c). 我们把高层次的网络层的特征称之为内容表示.

为了获得输入图像的风格表示, 我们使用一种最开始用于捕获纹理信息的一种特征空间. 这个特征空间在每一个网络层的过滤器的最顶上被构建出来. 通过加入很多层的特征相关性, 我们获得输入图像的静止的, 多尺度的表示, 这种表示捕获的是其纹理信息而不是全局ß布局信息.

Figure 1.

**CNN.** 一个给定的输入图像在 CNN 的每一个处理层次上表示为一组过滤后的图像. 随着层次处理过程中不同的过滤器的数量的逐渐增加, 通过一些降低取样的机制(e.g., 最大池化, 这种机制能够减少每一个神经层的单元数量), 过滤后的图像的大小会减少. 

**内容重构.** 我们能够通过只是用特定的层能够获取的网络层响应来重构图像, 这样就可以可视化 CNN 的每一个不同处理层次的信息. 我们从原始 VGG 网络的 a ~ e 层进行重构图像. 我们发现低层次的网络层重构出来的图像基本上比较完整(跟原始图像相比), 而在高层次的网络层, 具体的像素点有些就丢失了, 不过图像的高层次内容被保留了下来.

**风格重构.** 在原始 CNN 表示之上, 我们还构建了一种能够捕获输入图像的风格的新的特征空间. 风格表示通过计算 CNN 不同网络层中的不用特征之间的相关性. 我们从不同的 CNN 层子集的风格表示来重构输入图像的风格. 这可以在抛弃一些场景的全局排列的信息时以递增规模创建出匹配给定风格图像的风格的图像.

再次说明, 我们可以通过构建一张匹配给定风格图像的风格表示的图像来可视化的展现出这些构建于不同网络层的风格特征空间捕获的信息. 从风格特征中重构出输入图片的纹理化版本可以依据色彩和局部结构保留大体的外观。并且，随着层次化处理，局部图像结构的大小和复杂度都在逐渐增加，这种结果可以用接受域的大小和特征复杂度的增加来解释。我们将这种多尺度表示称之为风格表示。

本文最重要的发现就是在 CNN 中内容和风格的表示是分开的。也就是说，我们能够独立地操作这两种表示来生成新的，视觉上有意义的图像。为了演示这种发现，我们从两种源图像各自使用其内容和风格表示进行混合来生成新的图像。特别地，我们这里匹配一张描述德国 Neckarfront 的照片的内容表示，然后风格表示我们使用一些不同艺术阶段的很有名的画作。（Fig 2）

这些图像通过同时匹配照片的内容表示和这些画作各自的风格表示（查阅 Methods 部分了解更多）来合成。在保留原始照片的全局布局的同时，色彩和构成全局风格的局部结构就是由画作提供。算法高效渲染成画作的风格，其合成图像的外观跟画作很像，就算其内容显示的是照片的内容。

Fig 2.

组合照片内容和几副知名画作的风格的图像。这些图像通过同时匹配照片的内容表示和这些画作各自的风格表示（查阅 Methods 部分了解更多）来合成。原始照片取景于德国的一个小镇。为各自的生成图像提供风格的画作在左下角。

就像前面所提到的，风格表示是包含多个网络层的多尺度表示。Fig 2 中的图像是的风格表示是包含了整个网络层次的所有网络层。通过只包含一部分低层次的网络层，风格可以更加的局部化，这样可以得到不同的视觉体验（Fig 3 沿着行去观察）。当匹配的是更高层次的网络层，局部图像结构在逐渐增大的尺度上被匹配，这会使得视觉体验更加的流畅和连续。因此，匹配最高层网络层的风格表示的图像视觉上最吸引人（Fig 3 最后一行）

当然，图像内容和风格没法完全的解耦。在把一张图片的风格组合在另外一张图片中的时候，很难完美地同时匹配这两个限制。不过，我们把在图像合成的时候最小化的损失函数设计为含有内容和风格各自独立的两项限制。我们因此能够平滑地调节来强调是更多地考虑重构其内容还是其风格（Fig 3，顺着列去观察）。过重地强调风格会导致图像更加匹配画作，高效地得到一个材质化的版本，但是这会导致很难表现其照片的原来的内容（例如 Fig 3 的第一列）。对于每一对素材图片，我们可以自己权衡并调整内容和风格的权重来创造出视觉效果最吸引人的图片。

在此，我们展示这个人工神经系统能够将图像内容从风格风格开，这样可以将图像内容转换成其他任意图片的风格。我们还将展示组合多张画作的风格，使用任意选择的照片来创造新的艺术作品。在本文中，我们从高性能的从物体识别中训练得到的深度神经网络对应的特征得到图像内容和风格的神经表示。据我们所知，这是第一次展示从生活中的图片分离出风格的内容特征。在从内容中分离掉风格上，以往的工作都是估计具有更少复杂度的可感知输入上，例如不同手写体不同脸部照片或者是不同姿势的图片的普遍特征。


Fig 3.

使用画作 *Composition VII* 的风格的详细结果。这些行表示的是匹配越来越多的 CNN 层字集的风格表示的结果。我们发现在包含更高层次的网络层的风格特征之后由风格表示捕获的局部图像结构具有更大的大小和复杂度。这可以用在网络层次训练的时候其接受域和特征复杂度都一直在增加。这些列表示内容和风格重构之间不同的相关权重的影响。在每一列的索引位置的数字是更加匹配照片内容还是匹配画作风格的比率 $\frac{\alpha}{\beta}$ 。

在我们的演示中，我们将一张照片渲染成一组知名的画作的风格。这个问题一般是一个叫做非照片模拟渲染的计算机视觉分支所要处理的问题（最近研究的回顾可以从引用 14 了解）。从概念上讲， 最相近的一些有关方法都是使用材质转换来达到风格化转换的目的。然而，这些以往的方法主要依赖于直接操纵图像的像素表示的非参数技术（也就是提前没有作任何依赖于参数的假设，也不需要去求其参数）。相反的，使用用于训练物体识别的深度神经网络，我们能够将我们的操作放在显式地表示图像的高阶内容的特征空间上进行。

训练用于物体识别的深度神经网络的特征以前被用作风格识别，用来将艺术品按照艺术品被创造的时期来分类。此处，分类器在原始网络激活层的顶部被训练，我们把它们称为内容表示。我们推断像我们这种风格表示一样像稳定特征空间转换会取得比风格分类更加好的效果。

总的来说，我们的通过混合不同来源的内容和风格合成图像的方法提供了一种全新的，吸引人的工具来学习艺术，风格和不依赖内容的图像外观的感知机和神经表示。我们可以通过引入两种独立的，可感知的有意义的突变来源来设计新颖的刺激，这两种突变来源就是图像的外观和内容。我们可以预想此方法可以为很大一部分有关于可视化感知机的实验研究提供有力的工具，从心理物理学功能成像到电生理学的神经元记录。实际上，我们的工作提供了对神经表示如何独立地捕获图像的内容和它所在的风格的一种算法视角的理解。重要的是，风格表示的数学形式生成了一种关于图像外观的表示在单个神经元级别的明朗的，可测试的假设。风格表示简单地计算网络中不同类型的神经元的相关性。从神经元之间提取相关性是一种生物学上貌似可信的计算，例如，由所谓的主视觉加工系统的复杂细胞实现的计算。我们的结果认为进行在腹侧通路的不同处理阶段之间尽心类似于复杂细胞的那种计算可能是一种得到可视化输入的外观的不依赖内容的表示的思路。

总而言之，一个被训练用于处理生物视觉的一个核心计算任务，并且能够从允许吧图像内容从风格分隔开的图像表示自动学习的神经系统是真的很迷人。可能的解释是，当训练物体识别的时候，网络不得不对于所有包含物体标识的不同图像变成一个不变的模型。将不同图片因式分解为图像内容和它的外观对于这种任务相当的实用。因此，我们将内容从风格中抽象出来的能力，和创造以及享受艺术的能力是我们的视觉系统的强劲推断能力的一个主要信号。


### 方法
